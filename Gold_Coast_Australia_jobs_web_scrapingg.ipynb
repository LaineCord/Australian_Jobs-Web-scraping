```python
import pandas as pd
```

```python
#Bibliotecas
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
#Inicialização de um navegador automatizado utilizando o selenium
service = Service()
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(options=options)
```

```python
#Acesso a URL para extração dos dados
time.sleep(2)
url = 'https://au.indeed.com/jobs?q=&l=Gold+Coast+QLD&from=searchOnDesktopSerp&vjk=688130c39061ed9c'
driver.get(url)
```

# 1.Selecionando os itens

## 1.1. Titulos

```python
#Selecionando os titulos das vagas por pagina
titulos = driver.find_elements(By.CSS_SELECTOR, "h2.jobTitle a span")

for titulo in titulos:
    print(titulo.text)
```

## 1.3. Descrição da Vaga

```python
# Seleciona as descrições das vagas por pagina
descricoes = driver.find_elements(By.CLASS_NAME, "jobMetaDataGroup")


for descricao in descricoes:
    print(descricao.text)
```

# 2. Coletando os itens de todas as páginas

### 2.1. Looping para coletar todos os titulos de todas as páginas

```python
todos_titulos = []

while True:
    # Seleciona todos os títulos dos empregos na página atual
    titulos = driver.find_elements(By.CSS_SELECTOR, "h2.jobTitle a span")
    
    # Armazena os títulos extraídos
    for titulo in titulos:
        todos_titulos.append(titulo.text)
    
    # Tenta clicar no botão ">" se ele estiver disponível
    try:
        botao_proxima = driver.find_element(By.CSS_SELECTOR, "li.css-227srf a[data-testid='pagination-page-next']")
        botao_proxima.click()
        time.sleep(2)  # Aguarda para a próxima página carregar completamente

    except NoSuchElementException:
        # Se o botão ">" não estiver presente, encerra o loop

        break

for titulo in todos_titulos:
    print(titulo)
```

## 2.2. Looping para coletar localização

```python
todas_localizacoes = []

while True:
    # Seleciona todos os elementos de localização dos empregos na página atual
    localizacoes = driver.find_elements(By.CLASS_NAME, "company_location")
    
    # Armazena as localizações extraídas
    for localizacao in localizacoes:
        todas_localizacoes.append(localizacao.text)
        
    # Tenta clicar no botão ">" se ele estiver disponível
    try:
        botao_proxima = driver.find_element(By.CSS_SELECTOR, "li.css-227srf a[data-testid='pagination-page-next']")
        botao_proxima.click()
        time.sleep(2)  # Aguarda para a próxima página carregar completamente

    except NoSuchElementException:
        # Se o botão ">" não estiver presente, encerra o loop
        break

for localizacao in todas_localizacoes:
    print(localizacao)
```

## 2.3. Looping para coletar descrição

```python
todas_descricoes = []

while True:
    # Seleciona todos os elementos de descrição na página atual
    descricoes = driver.find_elements(By.CLASS_NAME, "jobMetaDataGroup")
    
    # Armazena as descrições extraídas
    for descricao in descricoes:
        todas_descricoes.append(descricao.text)
        
    # Tenta clicar no botão ">" se ele estiver disponível
    try:
        botao_proxima = driver.find_element(By.CSS_SELECTOR, "li.css-227srf a[data-testid='pagination-page-next']")
        botao_proxima.click()
        time.sleep(2)  # Aguarda para a próxima página carregar completamente

    except NoSuchElementException:
        # Se o botão ">" não estiver presente, encerra o loop
        break

for descricao in todas_descricoes:
    print(descricao)
```

## 2.4. DataFRAME com as informações coletadas

```python
dictDF = {'titulo': todos_titulos,
          'localizacao': todas_localizacoes,
          'descricao': todas_descricoes
         }

max_len = max(len(v) for v in dictDF.values())  # Encontrar o comprimento máximo

# Preencher listas curtas com None
for key, value in dictDF.items():
    while len(value) < max_len:
        value.append(None)


df = pd.DataFrame(dictDF)
df.head(20)
```
